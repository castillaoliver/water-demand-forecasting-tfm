Conversión de todos los archivos .txt en archivos .csv

```python
import glob
import os

archivos_txt = glob.glob(ruta_padrones + "/*.TXT")

lista_df = []

for archivo in archivos_txt:
    df = pd.read_csv(
        archivo,
        sep="\t",
        encoding="latin1",
        dtype=str
    )

    columnas_existentes = [c for c in columnas_utiles if c in df.columns]
    df = df[columnas_existentes].copy()

    nombre_archivo = os.path.basename(archivo)
    df["archivo_origen"] = nombre_archivo

    lista_df.append(df)

maestro_txt = pd.concat(lista_df, ignore_index=True)

maestro_txt.shape
```
