```python
import pandas as pd

ruta_maestro = "/content/drive/MyDrive/TFM/dataset_maestro_completo.csv"

maestro = pd.read_csv(
    ruta_maestro,
    sep=";",
    encoding="utf-8-sig",
    dtype=str
)

maestro.shape
```

Creación del dataframe con variables objetivo

```python
variables_modelo = [
    'cliente',
    'periodo',
    'fec_lec_ant_1',
    'fec_lec_act_1',
    'dif_dias_1',
    'm3_cons_1',
    'n_usuarios',
    'num_personas',
    'ruta',
    'des_tip_cli',
    'des_sub_tip_cli',
    'tipo_inc_1',
    'incidencia_1',
    'archivo_origen'
]

df_modelo = maestro[variables_modelo].copy()

df_modelo.shape
```

```python
df_modelo.head()
```

Convertir a numérico

```python
df_modelo["m3_cons_1"] = (
    df_modelo["m3_cons_1"]
    .str.replace(",", ".", regex=False)
    .astype(float)
)
```

```python
df_modelo.head()
```

Convertir fechas

```python
df_modelo["fec_lec_ant_1"] = pd.to_datetime(
    df_modelo["fec_lec_ant_1"],
    dayfirst=True,
    errors="coerce"
)

df_modelo["fec_lec_act_1"] = pd.to_datetime(
    df_modelo["fec_lec_act_1"],
    dayfirst=True,
    errors="coerce"
)
```

```python
df_modelo.head()
```

Creación de variables temporales

```python
df_modelo["anio"] = df_modelo["fec_lec_act_1"].dt.year
df_modelo["mes"] = df_modelo["fec_lec_act_1"].dt.month
df_modelo["bimestre"] = ((df_modelo["mes"] - 1) // 2) + 1
```

Consumo día

```python
df_modelo["dif_dias_1"] = (
    df_modelo["dif_dias_1"]
    .astype(str)
    .str.replace(",", ".", regex=False)
)

df_modelo["dif_dias_1"] = pd.to_numeric(
    df_modelo["dif_dias_1"],
    errors="coerce"
)
```

```python
df_modelo["m3_dia"] = df_modelo["m3_cons_1"] / df_modelo["dif_dias_1"]
```

Corrección final del df

```python
# Copia de trabajo
df_v2 = df_modelo.copy()

# Eliminar columnas que no aportan al modelado
columnas_eliminar = [
    "num_personas",
    "archivo_origen",
    "pobl_sumi",
    "cod_pobl_sumi"
]

df_v2 = df_v2.drop(
    columns=columnas_eliminar,
    errors="ignore"
)

# Eliminar registros sin días válidos de lectura
df_v2 = df_v2[df_v2["dif_dias_1"] > 0].copy()

# Recalcular consumo medio diario
df_v2["m3_dia"] = df_v2["m3_cons_1"] / df_v2["dif_dias_1"]

# Guardar dataset limpio
ruta_salida_v2 = "/content/drive/MyDrive/TFM/dataset_modelo_v2.csv"

df_v2.to_csv(
    ruta_salida_v2,
    sep=";",
    index=False,
    encoding="utf-8-sig"
)

print("Dataset v2 creado correctamente")
print(df_v2.shape)
```

comprobaciones

```python
# Tamaño del dataset limpio
print("Tamaño df_v2:")
print(df_v2.shape)

# Comprobar días de lectura
print("\nRegistros con dif_dias_1 <= 0:")
print((df_v2["dif_dias_1"] <= 0).sum())

# Consumos negativos
print("\nConsumos negativos:")
print((df_v2["m3_cons_1"] < 0).sum())

# Tipos de cliente
print("\nTipos de cliente:")
print(df_v2["des_tip_cli"].value_counts())

# Incidencias
print("\nTipos de incidencia:")
print(df_v2["tipo_inc_1"].value_counts())

# Periodos disponibles
print("\nNúmero de periodos:")
print(df_v2["periodo"].nunique())

print("\nRegistros por periodo:")
print(df_v2["periodo"].value_counts().sort_index())

# Consumo total por bimestre
consumo_bimestre = (
    df_v2
    .groupby("periodo")["m3_cons_1"]
    .sum()
    .sort_index()
)

print("\nConsumo total por bimestre:")
print(consumo_bimestre)

# Número de registros por bimestre
abonados_bimestre = (
    df_v2
    .groupby("periodo")["cliente"]
    .count()
    .sort_index()
)

print("\nNúmero de registros por bimestre:")
print(abonados_bimestre)
```
